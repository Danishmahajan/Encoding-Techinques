## Ordinal Number OR Label Encoding

<b>Label Encoding refers to converting the labels into numeric form so as to convert it into the machine-readable form. Machine learning algorithms can then decide in a better way on how those labels must be operated. </b>

```python
import datetime
```

```python
today_date=datetime.datetime.today() # getting the today date
```

```python
today_date
```

```python
# timedelta(3) subtract 3 days from today's day
today_date-datetime.timedelta(3)
```

```python
# List Comprehension
# I create a list of dates with 20 days difference from today
days=[today_date-datetime.timedelta(x) for x in range(0,20)]
days
```

```python
# making the days into a new dataframe
import pandas as pd
data=pd.DataFrame(days)
data
```

```python
data.columns=["Day"] # changing the column name
```

```python
data
```

```python
# extract the week day name
#data['day_of_week'] = data['Day'].dt.weekday_name
                         #or
data['weekday']=data['Day'].dt.strftime("%A")
data.head()
```

```python
#making a dictionary for maping 
dict={'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7}
```

```python
dict
```

```python
# making a new encoded feature that will maps the weekday name  
data['weekday_ordinal_encod`ed']=data['weekday'].map(dict)
```

```python
data
```
