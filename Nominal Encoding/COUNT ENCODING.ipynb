# COUNT ENCODING


<b>Count or frequency encoding: Replace the categories by the count of the observations that show that category in the dataset. Similarly, we can replace the category by the frequency -or percentage- of observations in the dataset.</b>

```python
import pandas as pd
df= pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
df.head() 
```

```python
#removing the columns name
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()
```

```python
# Selecting the categorical Features that i want to perform
columns=[1,3,5,6,7,8,9,13]
```

```python
#Changing the train_set that have our selectd features
train_set=train_set[columns]
train_set
```

```python
# Adding the columns Name To train _set
train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
#printing no. of unique labels in features
for feature in train_set.columns[:]:
    print(feature,":",len(train_set[feature].unique()),'labels')
```

```python
# Printing the value  counts of country feature
train_set['Country'].value_counts()
```

```python
# Adding a value_counts of country Feature to Dictionary
country_map=train_set['Country'].value_counts().to_dict()
country_map
```

```python
#map the country featyre with the count dict 
train_set['Country']=train_set['Country'].map(country_map)
train_set.head(20)
```

##### Now You can seee that the country name are encoded with value_counts

## Advantages
1.Easy To Use<br>
2.Not increasing feature space<br>

## Disadvantages
It will provide same weight if the frequencies are same
